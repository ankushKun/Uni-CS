**Plot multiple lines**

```python
month  = ["Jan", "Feb", "Mar","Apr", "May"]
SBI = [515.00, 502.70, 513.00, 512.00, 510.95]
PNB = [42.00, 41.55, 42.25, 42.20, 41.80]
plt.plot(month, SBI, 'g', label='SBI', linewidth=2)
plt.plot(month, PNB, 'b', label='PNB', linewidth=2)
plt.title('Top Banking Sector in India by Market Capitalization')
plt.ylabel('Share Price')
plt.xlabel('Month')
plt.legend() #Adding Legends
plt.show()
```

**Bar Graph using Title and legend**

```python
month  = ["Jan", "Feb", "Mar","Apr", "May"]
SBI = [515.00, 502.70, 513.00, 512.00, 510.95]
PNB = [42.00, 41.55, 42.25, 42.20, 41.80]
# First bar plot for SBI share
plt.bar(month, SBI, label="SBI", color='green')
# Second bar plot for PNB share
plt.bar(month, PNB, label="PNB", color='cyan')
plt.xlabel('Banks Share')
plt.ylabel('Month')
plt.title('Banking Sector Market Capitalization')
plt.legend()
plt.show()
```

**We will use the numpy for axis**

```python
import numpy as np 
import matplotlib.pyplot as plt 
month  = ["Jan", "Feb", "Mar","Apr", "May"]
SBI = [515.00, 502.70, 513.00, 512.00, 510.95]
PNB = [42.00, 41.55, 42.25, 42.20, 41.80]
#Here we will define the x-axis
X_axis = np.arange(len(month))
#-------------------------------#
# First bar plot for SBI share
plt.bar(X_axis-0.1, SBI, label="SBI", color='green')
# Second bar plot for PNB share
plt.bar(X_axis+0.1, PNB, label="PNB", color='cyan')
plt.xticks(X_axis, month)
plt.xlabel('Month')
plt.ylabel('Banks Share Price')
plt.title('Banking Sector Market Capitalization')
plt.legend()
plt.show()
```

**Pie Chart**

```python
# Creating dataset
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']
  
data = [23, 17, 35, 29, 12, 41]
plt.pie(data, labels=cars)
plt.show()
```

```python
# Creating dataset
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']
data = [23, 17, 35, 29, 12, 41]
# Use autopct to display the percent
# value using Python string formatting.
plt.pie(data, labels=cars, autopct='%1.1f%%') 
plt.show()
```

**subplots()**

```python
# x-axis and y-axis values for plot 1
month  = ["Jan", "Feb", "Mar","Apr", "May"]
PNB = [42.00, 41.55, 42.25, 42.20, 41.80]
# x-axis and y-axis values for plot 2
month  = ["Jan", "Feb", "Mar","Apr", "May"]
SBI = [515.00, 502.70, 513.00, 512.00, 510.95]
plt.subplot(1, 2, 1)
plt.plot(month,PNB)
plt.title("PNB")
plt.subplot(1, 2, 2)
plt.bar(month, SBI, label="SBI", color='green')
plt.title("SBI")
plt.show()
```
