# Get the list of datasets available in pycaret

```python
from pycaret.datasets import get_data
datasets = get_data("index")
```

# Loading Dataset - Loading dataset from pycaret

```python
bds = get_data("pokemon")
# bds.to_csv("to_csv.exmple.csv")
```

# Classification

```python
# Random Forest
from pycaret.classification import create_model, setup, plot_model

setup_data = setup(data=bds, target="Legendary")
rf_model = create_model('rf')
plot_model(rf_model, plot='confusion_matrix')
```
